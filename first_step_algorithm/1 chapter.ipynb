알고리즘에 사용되는 절차에는 기본적으로 세가지가 있다.
1. 순차구조
2. 선택구조
3. 반복구조


아무리 복잡한 알고리즘이라도 이 세가지 절차의 조합일 뿐이다. 이 세가지만 잘 알면 대부분의 알고리즘을 작성할 수 있다.

# 순차구조

처음부터 순서대로 처리하는 절차

예를 들어 마트에서 두부를 사오라는 심부름을 받았다면 1. 마트를 가고 2. 두부를 사고 3. 집으로 돌아간다. 이런 순서가 있을것이다. 이를 순서대로 작성하면

1. 마트를 간다.
2. 두부를 산다.
3. 집으로 간다.

이렇게 위에서부터 적어야 한다. 컴퓨터는 위에 있는 코드 부터 실행하니까 ^^

아무튼 이걸 순차구조라고 하고 인간이라면 너무나 익숙한 구조라서 공부할 필요도 없다.
겉옷을 입고 그 위에 상의를 입지 않는것 처럼 말이다.

# 선택구조

조건식으로 판단해 실행할 처리를 전환하는 절차

만약 마트를 갔는데 두부가 없다면 어찌할 것인가. 인간은 전화를 하거나 다른 판단을 하겠지만 컴퓨터는 오류가 난다.

그래서 조건 판단을 하도록 코드를 짜야한다. 1. 마트를 간다 2. 두부가 있나?  2-1. 두부가 있으면 두부를 산다 2-3. 두부가 없다면 아무것도 사지 않는다 3. 집으로 온다.

중요한 것은 일단 마트를 가야 조건 판단을 할 수 있게 해야하고 조건판단이 끝나면 집으로 돌아오는 마무리도 지어야 한다.

선택구조는 조건에 따라 처리가 나누어 지므로 분기구조라고도 하며 조건판단을 실시하는 문장을 조건식이라고 한다.

선택구조를 사용한 알고리즘은 급격하게 변화한다. 그리고 선택구조를 사용하지 않는 알고리즘은 거의 없다. 선택구조를 올바르게 사용할 수 있는게 중요포인트!
